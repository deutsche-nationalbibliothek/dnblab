```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
from lxml import etree
import unicodedata
```

```python
places = pd.read_csv("places.csv", encoding="utf-8")
places = places[:10]
places
```

```python
placenames = places.pubplace.to_list()
#print(placenames)
```

```python
def get_data(query):
    
    base_url = "https://services.dnb.de/sru/authorities"
    params = {'recordSchema' : 'MARC21-xml',
          'operation': 'searchRetrieve',
          'version': '1.1',
          'maximumRecords': '10',
          'query': 'cod=gik and ' + query
         }
    
    r = requests.get(base_url, params=params)
    xml = soup(r.content, features="xml")    
    #number = xml.find_all("numberOfRecords")[0].text  
    records = xml.find_all('record', {'type':'Authority'})
    
    
    #result = xml
    
    return(xml)
```

```python
result = [get_data(place) for place in placenames]
#output = pd.Dataframe(result)
#output
```
