```python
import pandas as pd
import unicodedata
```

```python
df = pd.read_csv("exilarchiv-monografien-mit-geoloc-v6-corr2.csv", sep=';', encoding="utf-8")
df_cord = pd.read_csv("exilarchiv_monografien-mit-geoloc_v5.csv", encoding="utf-8")
print(len(df))
print(len(df_cord))
```

```python
#df = df.rename(columns={'idn,Satzart,Haupttitel,Verlagsname,Erscheinungsort,Formangabe.online,author.name,author.vorname,author.idn,sprache.text,sprache.original,Erscheinungsjahr,idn.Druckausg,idn.reihe.a,Satzart.druck,Erscheinungsland.druck,Formangabe.druck,Sachgruppe,Geschlecht,Ersch.ort.no.brack,city,state,lat,long,Formangabe.bez.druck,Formangabe.bez.online': 'all'})
#df = df.rename(columns={'Ersch.ort.no.brack': 'pubplace'})
df = df.drop(['lat', 'long'], axis=1)
```

```python
df_cord = df_cord.rename(columns={'Ersch.ort.no.brack': 'pubplace'})
df_cord = df_cord[['pubplace', 'lat', 'long']]  # filter by pubplace (brackets already cleared)
#df_cord = df_cord[['Erscheinungsort', 'lat', 'long']]   # filter by Erscheinungsort (with brackets)
print(len(df_cord))
```

```python
df_cord = df_cord.drop_duplicates()
print(len(df_cord))
```

```python
# Remove white spaces at end of column "pubplace": 
def normalize(text): 
    new = unicodedata.normalize('NFC', text)
    return new 

df_cord.pubplace = df_cord.pubplace.apply(normalize)
df_cord.pubplace = df_cord.pubplace.str.strip('""')
```

```python
df_cord = df_cord.drop_duplicates()
print(len(df_cord))
```

```python
#df_merge = pd.merge(df, df_cord, on=['Erscheinungsort'], how="left")  # old version
df_merge_pubplace = df.merge(df_cord, on=['pubplace'], how="left")
```
