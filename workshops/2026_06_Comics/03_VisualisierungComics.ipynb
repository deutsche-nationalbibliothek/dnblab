```python
# Importieren der für die Visualisierungen nötigen Bibliotheken
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_file = "Comics-digitalAuthor&pub_yearBereinigt.csv"
```

```python
comics = pd.read_csv(csv_file, dtype={"pub_year": str})
```

```python
comics
```

```python
comics["pub_year"].value_counts()
```

```python
# Häufigkeiten der Erscheinungsdaten zählen
date_counts = comics["pub_year"].value_counts()

# Top 20 häufigsten Erscheinungsdaten
top_dates = date_counts.head(20)

# DataFrame für seaborn erstellen
df_top_dates = top_dates.reset_index()
df_top_dates.columns = ['pub_year', 'Anzahl']

# Horizontales Balkendiagramm (besser lesbar bei langen Datumsangaben)
plt.figure(figsize=(20, 6))
sns.barplot(
    data=df_top_dates,
    y='pub_year',
    x='Anzahl',
    hue='pub_year',  # So wird die Warnung vermieden
    palette="Blues_r",        # Farbpalette (r = reverse)
    legend=False              # Legende ausblenden
)
plt.title('Top 20 häufigste Erscheinungsdaten')
plt.xlabel('Anzahl der Veröffentlichungen')
plt.ylabel('Erscheinungsdatum')
plt.show()
```

```python
# Top 20 Autoren Darstellung Balkendiagramm
top_authors = comics["author"].value_counts().head(20)

# Dataframe erstellen (für Seaborn)
df_top = top_authors.reset_index()
df_top.columns = ['Autor', 'Beiträge']  

# Vertikales Balkendiagramm mit Beschriftung, definieren der Werte der x- und y-Achse
plt.figure(figsize=(20, 6))
sns.barplot(
    data=df_top,
    x='Autor',
    y='Beiträge',
    hue='Autor',  
    palette="viridis",
    legend=False  # Legende ausblenden, weil sie sonst doppelt wäre
)
#Beschriftung des Diagramms und Ausrichtung festlegen
plt.title('Top 20 Hauptautoren nach Anzahl der Beiträge')
plt.xlabel('Autor')
plt.ylabel('Anzahl der Beiträge')
plt.xticks(rotation=45)
plt.show()
```

```python
# Häufigkeiten der Autoren zählen
author_counts = comics["author"].value_counts()

# Top 20 Autoren auswählen
top_authors = author_counts.head(20)

# DataFrame erstellen (für seaborn)
df_top_authors = top_authors.reset_index()
df_top_authors.columns = ['Autor', 'Anzahl']

# Horizontales Balkendiagramm mit Beschriftung
plt.figure(figsize=(20, 6))

# Barplot erstellen
ax = sns.barplot(
    data=df_top_authors,
    y='Autor',
    x='Anzahl',
    hue='Autor',           
    palette="viridis",     
    legend=False           # Legende ausblenden
)

# Beschriftung der Balken (Zahlen auf den Balken)
for p in ax.patches:
    width = p.get_width()
    # Text rechts neben dem Balken platzieren
    ax.text(width + 0.05, p.get_y() + p.get_height()/2, f'{width:.0f}', 
            ha='left', va='center', fontsize=10, color='black')

# Titel und Beschriftungen
plt.title('Top 20 Hauptautoren nach Anzahl der Beiträge', fontsize=14, fontweight='bold')
plt.xlabel('Anzahl der Beiträge', fontsize=12)
plt.ylabel('Autor', fontsize=12)

# Layout anpassen (für bessere Lesbarkeit)
plt.tight_layout()

# Anzeigen
plt.show()
```
