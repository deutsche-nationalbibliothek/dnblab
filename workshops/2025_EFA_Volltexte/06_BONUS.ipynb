# BONUS-Notebook Metadatenanalyse

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("Studentenbewegung_free_pdf.csv", encoding="utf-8")
df
```

```python
# Häufigkeiten der Jahre zählen
year_counts = df['year'].value_counts().reset_index()
year_counts.columns = ['year', 'count']

# Barchart erstellen
fig = px.bar(year_counts, x='year', y='count', title='Häufigkeiten der Publikationsjahre')
fig.show()
```

```python
# Häufigkeiten der einzelnen Autoren zählen
creator_counts = df['author'].value_counts().reset_index()
creator_counts.columns = ['author', 'count']

#top_50_creators = creator_counts.head(50)

# Barchart erstellen
fig2 = px.bar(creator_counts, x='author', y='count', title='Häufigkeiten der Autoren')
fig2.show()
```

```python
top_20_creators = creator_counts.head(20)

# Piechart erstellen
fig3 = px.pie(top_20_creators, names=top_20_creators['author'], values=top_20_creators['count'], title='Top 20 Autoren')
fig3.show()
```
