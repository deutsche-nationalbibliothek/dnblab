## Herangehensweise der Datenbereinigung

Zunächst Publikationsdatum aus Datafield 264 $a ziehen: 

```python
#check for dates: 

for record in records: 
    xml = ET.fromstring(str(record))
    date1 = xml.xpath(".//datafield[@tag = '264']/subfield[@code = 'c']")
    if date1:
        print(date1[0].text)
    else:
        print("N/A")
```

Es fällt auf, dass bereits beim dritten Record kein Datum gefunden wurde. Warum ist das so? Hierzu oben noch einmal Treffer Nr. 3 (also [2]) anzeigen lassen und nachschauen: [Sprung zur beispielhaften Anzeige](#Beispielhafte-Anzeige-des-ersten-(0.)-Treffers.) 


Beim Nachschauen sollte aufgefallen sein, dass tatsächlich kein Datafield 264 a existiert, dafür aber ein Publikationsdatum im Hochschulschriftenvermerk (Datafield 502 a) zu finden ist. Daher wird der Code jetzt so angepasst, dass erst nachgeschaut wird, ob es ein entsprechendes 264-Feld gibt, falls dies nicht exisitert, ob es dann ein 502-Feld gibt, und erst wenn beide Felder nicht vorhanden sein sollten, wird ein "N/A"-Eintrag vorgenommen: 


```python
# modify and check more fields for date:

for record in records: 
    xml = ET.fromstring(str(record))
    date1 = xml.xpath(".//datafield[@tag = '264']/subfield[@code = 'c']")
    date2 = xml.xpath(".//datafield[@tag = '502']/subfield[@code = 'a']")
    if date1:
        print(date1[0].text)
    elif date2: 
        print(date2[0].text)
    else:
        print("N/A")
```

Nun sind zwar überall Einträge, allerdings teilweise mit viel Text, obwohl hier eigentlich nur die Jahreszahl interessant ist und später benötigt wird. Daher kann man den Code noch einmal anpassen. Da allen Feldern mit Hochschulschriftenvermerk gemeinsam ist, dass die Jahreszahl der Publikation am Ende des Feldes steht, reicht es also, die jeweils letzten 4 Zeichen des Hochschulschriftenvermerks zu behalten, der Rest wird abgeschnitten: 

```python
# modify and only keep last 4 digits from dissertation statement:

for record in records: 
    xml = ET.fromstring(str(record))
    date1 = xml.xpath(".//datafield[@tag = '264']/subfield[@code = 'c']")
    date2 = xml.xpath(".//datafield[@tag = '502']/subfield[@code = 'a']")
    if date1:
        print(date1[0].text)
    elif date2: 
        date2 = date2[0].text
        date2 = date2[-4:]       # Hier wird die Eingrenzung auf die letzten 4 Zeichen des Textes vorgenommen
        print(date2)
    else:
        print("N/A")
```

Das sieht schon deutlich besser aus! :) Zur Kontrolle kann man noch einmal durch die Ergebnisse scrollen, da nun optisch alle Einträge, die kürzer oder länger als vier Zahlen sind, deutlich herausstechen sollten. Und siehe da: Es gibt noch einige wenige Einträge, bei denen die Jahreszahl in eckigen Klammern steht. Auch das soll nun noch schnell bereinigt werden. Hierfür werden einfach alle [] entfernt: 

```python
# clean data: 

for record in records: 
    xml = ET.fromstring(str(record))
    date1 = xml.xpath(".//datafield[@tag = '264']/subfield[@code = 'c']")
    date2 = xml.xpath(".//datafield[@tag = '502']/subfield[@code = 'a']")
    if date1:
        date1 = date1[0].text
        date1 = date1.strip("[]")   # Entfernen der eckigen Klammern
        print(date1)
    elif date2: 
        date2 = date2[0].text
        date2 = date2[-4:]
        print(date2)
    else:
        print("N/A")
```

## Die fertige Funktion

Der Code, der alle Schritte der eben vorgenommenen Datenbereinigung enthält, sieht dann entsprechend wie im folgenden Codeblock dargestellt aus. Dabei wurde dieser nun noch in eine Funktion eingebettet und um einige wenige weitere Informationen ergänzt. 
So wird in der if-else-Schleife nun jeweils noch mit ausgegeben, an welcher Stelle die Jahreszahl gefunden wurde: Wird sie direkt am Anfang im 264 Feld gefunden, wird eine weitere Variable "source" gesetzt und dort "264" eingetragen. Existiert dieses Feld nicht, es wird aber ein Eintrag im 502-Feld gefunden, wird in die Variable source der Wert 502 eingetragen. Wird keines der beiden Felder gefunden, wird auch hier wieder "N/A" ausgegeben. 
Außerdem wird nun auch noch das Feld "controlfield" mit der Nummer 001 ausgegeben, welches die Identifikationsnummer (idn) des jeweiligen Datensatzes enthält. Dies dient vor allem dazu, sich bei Fehlern oder anderen Unklarheiten direkt den entsprechenden Datensatz ansehen zu können:

```python
# Extract data: 

def get_data(record): 
        
    xml = ET.fromstring(str(record))
    
    date1 = xml.xpath(".//datafield[@tag = '264']/subfield[@code = 'c']")
    date2 = xml.xpath(".//datafield[@tag = '502']/subfield[@code = 'a']")
    if date1:
        date1 = date1[0].text
        date = date1.strip("[]c 'f idnn'")
        source = "264"
    elif date2:
        date2 = date2[0].text
        date = date2[-4:]
        source = "502"
    else:
        date = "N/A"
        source = "N/A"
        
        
    idn = xml.xpath(".//controlfield[@tag = '001']")
    if idn:
        idn = idn[0].text
    
        
    results = {"date":date, "source":source, "idn":idn}
    
    return results
    
```

Mit dem folgenden Code wird dann jeder Treffer / Record einzeln der Funktion übergeben und nach den entsprechenden Feldern durchsucht, bevor die Ergebnisse dann in eine tabellarische Darstellung übertragen werden: 

```python
data = [get_data(record) for record in records]
df = pd.DataFrame(data)
df
```

Hier noch schnell prüfen, ob sich noch Einträge mit "N/A" in der Tabelle befinden: 

```python
# check for entries without a date: 

df.loc[df['date'] == "N/A"]
```

Und hiermit kann man schauen, ob es noch Einträge gibt, die länger als 4 Zeichen sind: 

```python
df.query("date.str.len() > 4")
```

## Erstellen der Visualisierung 

Mit Hilfe des folgenden Codes wird nun die Häufigkeit der jeweiligen Jahreszahlen im Datensatz dargestellt. Dabei wird für jeden Eintrag ein "Strich" gezeichnet, es handelt sich bei der Darstellung also quasi wirklich um einen Stapel an Einträgen. Dies lässt sich auch optisch schön an den Streifen der Balken erkennen. Auch zeigt das Mouseover jedesmal als Anzahl "1" an, auch wenn insgesamt natürlich fast 500 mal die Jahreszahl 2010 im Datensatz vorkommt: 

```python
import plotly.express as px

fig = px.bar(df, x='date')
fig.show()
```

Um das ganze nun graphisch etwas schöner dazustellen, bündeln wir die Anzeige und nutzen dafür ein Histogramm: 

```python
fig2 = px.histogram(df, x="date", barmode='group')
fig2.show()
```

Nun zeigta auch das Mouseover die Gesamtanzahl zur jeweiligen Jahreszahl. 

Auch kann die Anzeige natürlich noch auf- oder absteigend sortiert werden: 

```python
fig2 = px.histogram(df, x="date", barmode='group')
fig2.update_layout(barmode='stack', xaxis={'categoryorder':'category ascending'})
fig2.show()
```

Letzte Anmerkung: Hier sieht man gut, dass sich "Ausreißer" bei den Daten durch Visualisierungen gut erkennen lassen. Der Sprung von 1923 auf dann 1996 fällt direkt auf und lädt ein, die Richtigkeit dieser Angabe ggf. zu überprüfen. Mit folgendem Code lässt sich der entsprechende Eintrag auch noch einmal mit Quelle und IDN anzeigen: 

```python
df.query('date.str.contains("1923")')
```
